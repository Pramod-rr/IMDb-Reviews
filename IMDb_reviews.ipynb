```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import re
```

```python
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model = AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
def sentiment_score(df): #function to calculate sentiment score
    tokens = tokenizer.encode(df, return_tensors='pt')
    result = model(tokens)
    score = torch.softmax(result.logits, dim=1)
    return torch.sum(score * torch.arange(1, score.size(1) + 1)).item().__round__(2)
```

```python
print(f"{sentiment_score(df['review'].iloc[9])}/5") #test the function
```
